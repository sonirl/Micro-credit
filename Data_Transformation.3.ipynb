```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
data = pd.read_csv('data_cleaned_2.csv')
data.head()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
data.describe()
```

Now we have seen in the visualisation that our data is rightly skewed, so I will be using log transformation so that it will normalise our data and also reduce the skewness

```python
for i in data:
    print(i)
    print('skewness of data before transformation :', data[i].skew())
    print('skewness of data after transformation :', np.log1p(data[i]+1).skew())
    print('\n')
```

There are some columns which are negatively affected by log transformation, so I will be transforming selected columns only

```python
data['aon'] = np.log1p(data['aon']+1)
data['daily_decr30'] = np.log1p(data['daily_decr30']+1)
data['daily_decr90'] = np.log1p(data['daily_decr90']+1)
data['last_rech_date_ma'] = np.log1p(data['last_rech_date_ma']+1)
data['cnt_ma_rech30'] = np.log1p(data['cnt_ma_rech30']+1)
data['fr_ma_rech30'] = np.log1p(data['fr_ma_rech30']+1)
data['cnt_ma_rech90'] = np.log1p(data['cnt_ma_rech90']+1)
data['fr_ma_rech90'] = np.log1p(data['fr_ma_rech90']+1)
data['cnt_loans30'] = np.log1p(data['cnt_loans30']+1)
data['amnt_loans30'] = np.log1p(data['amnt_loans30']+1)
data['cnt_loans90'] = np.log1p(data['cnt_loans90']+1)
data['amnt_loans90'] = np.log1p(data['amnt_loans90']+1)
data['payback30'] = np.log1p(data['payback30']+1)
data['payback90'] = np.log1p(data['payback90']+1)
```

```python
data.skew()
```

```python
data.to_csv('log_transformed.csv')
```
