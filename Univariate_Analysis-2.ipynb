```python
#creating continous variable data
cont_data = data[['label','aon','daily_decr30','daily_decr90','rental30','rental90','last_rech_amt_ma','sumamnt_ma_rech30',
            'medianamnt_ma_rech30','medianmarechprebal30','sumamnt_ma_rech90','medianamnt_ma_rech90','medianmarechprebal90',
            'amnt_loans30','amnt_loans90','payback30','payback90']]
```

```python
fig, ax = plt.subplots(ncols=4 ,nrows=5 , figsize=(20,10))
index = 0
ax = ax.flatten()

for i,value in cont_data.items():
    sns.histplot(x=i, data=cont_data,hue='label' ,ax=ax[index])
    index += 1
plt.tight_layout(pad=0.5, h_pad= 1, w_pad= 0.6)
```

```python
for i in cont_data:
    sns.FacetGrid(cont_data,hue='label',size=6).map(sns.distplot,i).add_legend()
    print('Distribution of ', i)
    plt.xticks(rotation=90)
    plt.show()
```

From the above plots, we can see that when the value of variable is minimum or close to 0 then there are more number of defaulters, as the value increases the number of defaulters decreases.
