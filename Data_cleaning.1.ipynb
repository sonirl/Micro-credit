### DATA INFO

```python
data.info()
```

most of our variables are numeric, pdate variable needs to be changed to datetime format to extract useful values.

Unnamed: 0, pcircle, and  msisdn are not necessary for our prediction 
pcircle = it has only one value in all the observation, there is no variance so removing this will be a better approach.
Unnamed: 0 and msisdn are there only to locate the indivisual observation.

```python
data = data.drop(['Unnamed: 0','msisdn','pcircle'],axis=1)
#Extracting date and month from pdate column
data['pdate'] = pd.to_datetime(data['pdate'])
data['date'] = data['pdate'].dt.day
data['month'] = data['pdate'].dt.month
data.drop('pdate',axis=1,inplace=True)
```

### Statistical description

```python
data.describe()
```

from the above discription we see that almost our data is imbalanced, we have more counts of 1 than 0 in or target variable.
Almost every independent variable is highly skewed to right and has outliers. 
aon shows negative value at minimum as age (in terms of days) cannot be negative, daily_decr30 and daily_decr90 also shows genative value at minimum that too needs to be corrected.
last_rech_date_ma and last_rech_date_da - minimum value also here should be corrected.

I will 1st convert aon, daily_decr30,  daily_decr90, last_rech_date_ma and last_rech_date_da into absolute values to remove the negative value.

I will replace the values which are more than (Q3 + 1.5(IQR)) with Q3 + 1.5(IQR) and in some comlumns where  values are less than  (Q1-1.5(IQR)) with Q1-1.5(IQR)

```python
#cleaning aon
data['aon']= abs(data['aon'])
data['aon'] = np.where(data['aon']>data['aon'].quantile(0.75) + 1.5*(data['aon'].quantile(0.75)-data['aon'].quantile(0.25)),data['aon'].quantile(0.75) + 1.5*(data['aon'].quantile(0.75)-data['aon'].quantile(0.25)),data['aon'])
data['aon'].describe()
```

```python
#cleaning daily_decr30 and daily_decr90
data['daily_decr30']= abs(data['daily_decr30'])
data['daily_decr90']= abs(data['daily_decr90'])

data['daily_decr30'] = np.where(data['daily_decr30']>data['daily_decr30'].quantile(0.75) + 1.5*(data['daily_decr30'].quantile(0.75)-data['daily_decr30'].quantile(0.25)),data['daily_decr30'].quantile(0.75) + 1.5*(data['daily_decr30'].quantile(0.75)-data['daily_decr30'].quantile(0.25)),data['daily_decr30'])
data['daily_decr90'] = np.where(data['daily_decr90']>data['daily_decr90'].quantile(0.75) + 1.5*(data['daily_decr90'].quantile(0.75)-data['daily_decr90'].quantile(0.25)),data['daily_decr90'].quantile(0.75) + 1.5*(data['daily_decr90'].quantile(0.75)-data['daily_decr90'].quantile(0.25)),data['daily_decr90'])
print(data['daily_decr30'].describe())
print('*****')
print(data['daily_decr90'].describe())
```
